```python
# ----------------------------------------
# IMPORT LIBRARIES
# ----------------------------------------
import pandas as pd
import numpy as np


# ----------------------------------------
# CREATE BASIC DATA STRUCTURES
# ----------------------------------------
labels = ['a', 'b', 'c']              # Custom index labels

my_list = [10, 20, 30]               # Python list

my_array = np.array([10, 20, 30])    # NumPy array

my_dict = {"name": "sam", "age": 20} # Dictionary (key-value pairs)


# ----------------------------------------
# SERIES FROM LIST
# ----------------------------------------
print("Series from List (default index)")
print(pd.Series(my_list))            # Default index (0,1,2)
print()


print("Series from List (custom index)")
print(pd.Series(my_list, index=labels))  # Custom index (a,b,c)
print()


# ----------------------------------------
# SERIES FROM NUMPY ARRAY
# ----------------------------------------
print("Series from NumPy Array")
print(pd.Series(my_array, index=labels))  # Works same as list
print()


# ----------------------------------------
# SERIES FROM DICTIONARY
# ----------------------------------------
print("Series from Dictionary")
print(pd.Series(my_dict))            # Keys become index automatically
print()
```

```python
import pandas as pd

# Create first DataFrame
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["A", "B", "C"]
})

# Create second DataFrame
df2 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Salary": [10000, 20000, 30000]
})
```

```python
df1
```

```python
df2
```

```python
# Merge both DataFrames on common column "ID"
# Default = INNER JOIN --> (only matching values kept)
result = pd.merge(df1, df2, on="ID")    # by default -> how = "inner"

# Print result
print("MERGE OUTPUT:")
print(result)
```

```python
result = pd.merge(df1, df2, on="ID" , how="outer") # how = "outer" -> take all the id's of both dataset
print("MERGE OUTPUT:")
print(result)
```

```python
# Create DataFrames

df3 = pd.DataFrame({
    "A": [1, 2],
    "B": [3, 4]
})

df4 = pd.DataFrame({
    "A": [5, 6],
    "B": [7, 8]
})
```

```python
df3
```

```python
df4
```

```python
# 🔸 Row-wise concatenation (default axis=0)
# Stacks df2 below df1

row_concat = pd.concat([df1, df2], axis = 0)

print("\nROW CONCAT:")
print(row_concat)

# 🔸 Column-wise concatenation (axis=1)
# Adds df2 as new columns beside df1

col_concat = pd.concat([df1, df2], axis=1)

print("\nCOLUMN CONCAT:")
print(col_concat)
```

```python
# Create DataFrames
df5 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["A", "B", "C"]
})

df6 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Salary": [10000, 20000, 30000]
})
```

```python
df5
```

```python
df6
```

```python
# Set "ID" as index for both DataFrames

df5.set_index("ID", inplace=True)
df6.set_index("ID", inplace=True)

# Join based on index

# Default = LEFT JOIN (keeps all df5 values)
result = df5.join(df6)    # how -> inner/outer works also in the join 

print("\nJOIN OUTPUT:")
print(result)
```

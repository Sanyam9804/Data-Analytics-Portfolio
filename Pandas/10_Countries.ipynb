```python
import pandas as pd
import numpy as np 

df = pd.read_csv('Countries.csv')
df.head()
```

```python
# WHICH COUNTRY HAS HIGHEST POPULATION
df[df['population'] == df['population'].max()] 
```

```python
df[df['population'] == df['population'].max()]['country']
```

```python
# WHAT IS THE CAPITAL OF COUNTRY WHICH HAVE HIGHEST POPULATION
df[df['population'] == df['population'].max()]['capital_city']
```

```python
# WHICH COUNTRY HAS LEAST POPULATION
df[df['population'] == df['population'].min()]['country']
```

```python
# WHAT IS THE CAPITAL OF COUNTRY WHICH HAVE LEAST POPULATION
df[df['population'] == df['population'].min()]['capital_city']
```

```python
# TOP 5 COUNTRY WHICH HAVE HIGHEST DEMOCRATIC SCORE
df.nlargest(5,'democracy_score')
```

```python
df.nlargest(5,'democracy_score')['country']
```

```python
# HOW MANY TOTAL REGIONS ARE THERE
count = df['region'].count()
print(count)
```

```python
# HOW MANY COUNTRIES LIE IN EASTER EUROPE REGION
df[df['region'] == "Eastern Europe"].count()
```

```python
# WHO IS THE LEADER OF 2nd HIGHEST POPULATED COUNTRY
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
# how many countries are there whoes political leaders are unknown
df[df['political_leader'].isna()]['country'].count()
```

```python
#how many country have Republic in their full name
count = 0
def counting(txt):
    global count 
    if 'republic' in txt.lower():
        count+= 1
    return txt
df['country_long'] = df['country_long'].apply(counting)
print(count)
```

```python
#which country in african region has highest population
africa_df = df[df['continent'] == 'Africa']
```

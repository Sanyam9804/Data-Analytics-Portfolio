```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Name": ["A", "B", "C", "D", "E"],
    "Department": ["IT", "HR", "IT", "HR", "IT"],
    "Salary": [50000, 40000, 60000, 45000, 70000],
    "Age": [25, 30, 28, 35, 40]
})

print(df)
```

```python
# Group data 

# I WANT TO DIVIDE CATEGORY ON BASIS OF DEPT.

Data = df.groupby('Department')

for i,j in Data:
    print(i)
    print(j)
```

```python
# I WANT TO TOTAL SALARY DEPT. WISE


Data = df.groupby('Department')['Salary'].sum()
Data
```

```python
# I WANT AVG AGE DEPT. WISE

Data = df.groupby('Department')['Age'].mean()
Data
```

```python
# I WANT TO GROUBY MULTIPLE COLUMN

Data = df.groupby(['Department','Age'])['Salary'].sum()
Data
```

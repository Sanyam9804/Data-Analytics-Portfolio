```python
import pandas as pd
import numpy as np
```

```python
# Creating a DataFrame using dictionary

data = {
    "Name": ["Sam", "Ram", "Shyam"],
    "Age": [20, 25, 30],
    "Salary": [30000, 40000, 50000]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Creating a DataFrame using list

data = [
    ["Sam", 20, 30000],
    ["Ram", 25, 40000],
    ["Shyam", 30, 50000]
]

df2 = pd.DataFrame(data,columns=["Name", "Age", "Salary"]) 

print(df2)
```

```python
# Selecting single column
print(df["Name"])

print()

# Selecting multiple columns
print(df[["Name", "Salary"]])
```

```python
# Creating a new column using existing columns

df["Bonus"] = df["Salary"] * 0.10

print(df)
```

```python
# Dropping a column (Method 1: Reassign)

df.drop("Bonus", axis=1)   # axis=1 → column

# ----------------------------------------
# Alternative Method (Method 2: inplace)
# ----------------------------------------

# 
# df.drop("Bonus", axis=1, inplace=True)  
# # inplace=True → modifies original DataFrame

# print(df)
```

```python
print(df)
```

```python
# Selecting rows using index

print(df.loc[0])        # by label
print()

print(df.loc[[0,1]])        # by label multiple
print()

print(df.iloc[1])       # by position / index
```

```python
# Selecting subset of rows and columns

print(df.loc[0:1, ["Name", "Salary"]])   # label-based
print()

print(df.iloc[0:2, 0:2])                 # position-based
```

```python
print(df)
```

```python
# Filtering data based on condition

# df[(condition1) & (condition2)]

# AND -> & , OR -> |

print(df[(df["Salary"] > 35000) & (df["Age"] > 20)])
```

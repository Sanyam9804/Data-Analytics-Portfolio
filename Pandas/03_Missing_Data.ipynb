```python
import pandas as pd
import numpy as np

# Creating DataFrame with missing values

data = {
    "Name": ["Sam", "Ram", "Shyam", None],
    "Age": [20, np.nan, 30, 25],
    "Salary": [30000, 40000, np.nan, np.nan]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Check missing values 

df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any() # any() -> proivde how many null values in each column 
```

```python
df
```

```python
# Removing Missing Dataabs -> works on row basis

df.dropna()   # all the rows are drop which have null values



 
```

```python
df
```

```python
df.dropna(thresh=2)   # thresh = 2 -> means in row minimum two non zero element are there
```

```python
# Fill missing values 

df
```

```python
df.fillna(0)
```

```python
values = {"Name":"ZZZ","Age":000,"Salary":"XXXX"}
df.fillna(value=values)   # value -> is a keyword
```

```python
df
```

```python
# Fill values using mean of that columnabs

df.fillna(df.mean()) # while taking mean -> data should be in numbers
```

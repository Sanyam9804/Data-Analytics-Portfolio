```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

**BASICS**

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("LINE GRAPH")
plt.xlabel("X - AXIS")
plt.ylabel("Y - AXIS")
```

**SUBPLOT**

```python
plt.subplot(2,2,1) # 2,2 mean 2 row and 2 col ans 1 means the first box of the 2,2 matrix
plt.title("1 BOX")
plt.plot(x,y)

plt.subplot(2,2,2)
plt.title("2 BOX")
plt.plot(x**2,x**2)

plt.subplot(2,2,3)  
plt.title("3 BOX")
plt.plot(x**2,y**-2)

plt.subplot(2,2,4)
plt.title("4 BOX")
plt.plot(y,x)
```

**OBJECT ORIENTED WAY**

```python
fig = plt.figure(figsize=(10,4) , dpi=500)   # CANVAS , dpi = increase clearity of the graph , figsize -> length,width of the canvas
axis1 = fig.add_axes([0.1,0.1,1,1]) 
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
# Add a new set of axes (plot area) to the figure

# fig.add_axes([
#     0.1,  # left → distance from left side of figure (10%)
#     0.1,  # bottom → distance from bottom of figure (10%)
#     0.8,  # width → width of axes (80% of figure width)
#     0.8   # height → height of axes (80% of figure height)
# ])
```

**TYPES OF PLOT**

```python
# SCATTER PLOT -> DOT GRAPH
plt.scatter(x,y)
```

```python
# HISTOGRAM -> CONTAIN ONLY SINGLE DATA LIKE x or y 
plt.hist(y)
```

```python
# BOX -> CONTAIN ONLY SINGLE DATA LIKE x or y 
plt.boxplot(y)
```

**SAVING PLOTS**

```python
fig = plt.figure(figsize=(10,4) , dpi=500)  
axis1 = fig.add_axes([0.1,0.1,1,1]) 
axis1.plot(x,y)
plt.savefig("BasicPlot.png")
```
